```python
districts = {

# Punjab
"Amritsar": (31.63, 74.87),
"Barnala": (30.37, 75.55),
"Bathinda": (30.21, 74.95),
"Faridkot": (30.67, 74.76),
"Fazilka": (30.40, 74.03),
"Ferozepur": (30.92, 74.61),
"Gurdaspur": (32.04, 75.40),
"Hoshiarpur": (31.53, 75.91),
"Jalandhar": (31.33, 75.57),
"Kapurthala": (31.38, 75.38),
"Ludhiana": (30.90, 75.85),
"Mansa": (29.99, 75.40),
"Moga": (30.82, 75.17),
"Muktsar": (30.47, 74.52),
"Pathankot": (32.27, 75.65),
"Patiala": (30.34, 76.40),
"Rupnagar": (30.97, 76.53),
"Sangrur": (30.23, 75.84),
"Tarn Taran": (31.45, 74.93),

# Haryana
"Ambala": (30.37, 76.78),
"Bhiwani": (28.79, 76.13),
"Faridabad": (28.41, 77.32),
"Fatehabad": (29.52, 75.45),
"Gurugram": (28.46, 77.03),
"Hisar": (29.15, 75.72),
"Jhajjar": (28.61, 76.66),
"Jind": (29.31, 76.31),
"Kaithal": (29.80, 76.40),
"Karnal": (29.68, 76.99),
"Kurukshetra": (29.96, 76.87),
"Mahendragarh": (28.28, 76.15),
"Nuh": (28.10, 77.00),
"Palwal": (28.14, 77.33),
"Panchkula": (30.69, 76.85),
"Panipat": (29.39, 76.97),
"Rewari": (28.20, 76.62),
"Rohtak": (28.89, 76.59),
"Sirsa": (29.53, 75.02),
"Sonipat": (28.99, 77.02),
"Yamunanagar": (30.13, 77.28)

}
```

```python
for district, (lat, lon) in districts.items():

    url = (
        "https://power.larc.nasa.gov/api/temporal/daily/point"
        "?parameters=PRECTOTCORR,T2M_MAX,T2M_MIN,RH2M,WS2M"
        "&community=RE"
        f"&longitude={lon}"
        f"&latitude={lat}"
        "&start=20150101"
        "&end=20231231"
        "&format=CSV"
    )

    response = requests.get(url)

    file_path = f"data/raw/weather/{district}_weather.csv"

    with open(file_path, "w") as f:
        f.write(response.text)

    print("Saved:", district)

    time.sleep(1)

print("All districts downloaded!")
```
