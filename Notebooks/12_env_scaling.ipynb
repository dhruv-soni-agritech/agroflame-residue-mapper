# Notebook 12 — Environmental Feature Scaling

**Purpose:** Apply StandardScaler to the four environmental features so they have zero mean and unit variance, making them suitable for distance-based algorithms like K-Means and PCA.

**Input:** `Data/Processed/env_features.csv`  
**Output:** `Data/Processed/env_scaled.csv`, `Data/Processed/env_scaler.pkl`

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import pickle
```

```python
# Load feature table 
env = pd.read_csv('../Data/Processed/env_features.csv')
print(f"Shape: {env.shape}")
print(env.head())
print(f"\nDescriptive stats (raw):")
print(env.describe().round(2))
```

```python
# Scale features 
# StandardScaler subtracts the mean and divides by the standard deviation.
# This is required before K-Means (which uses Euclidean distance) and PCA
# so that high-magnitude features like fire_count don't dominate.
features = ['fire_count', 'residue', 'avg_temp', 'rainfall']
X        = env[features]

scaler   = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Store as DataFrame with '_scaled' suffix on column names
X_df = pd.DataFrame(X_scaled, columns=[f + '_scaled' for f in features])
X_df['district'] = env['district'].values

print("Scaled feature means (should all be ~0):")
print(X_df[[c for c in X_df.columns if '_scaled' in c]].mean().round(4).to_string())
print("\nScaled feature stds (should all be ~1):")
print(X_df[[c for c in X_df.columns if '_scaled' in c]].std().round(4).to_string())
print()
print(X_df.head())
```

```python
# Save scaled data and scaler 
X_df.to_csv('../Data/Processed/env_scaled.csv', index=False)

with open('../Data/Processed/env_scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

print("Saved: Data/Processed/env_scaled.csv")
print("Saved: Data/Processed/env_scaler.pkl")
```
